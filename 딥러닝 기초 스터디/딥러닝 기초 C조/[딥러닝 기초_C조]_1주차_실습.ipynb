```python
import numpy as np
```

AND 게이트 구현하기 : 가중치(Weight)

```python
def AND_Gate_Weight(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  result = x1*w1 + x2*w2
  if result<= theta: 
    return 0
  elif result > theta:
    return 1

print(AND_Gate_Weight(0,0),AND_Gate_Weight(1,0),AND_Gate_Weight(0,1),AND_Gate_Weight(1,1), sep = ',')
```

AND 게이트 구현하기 : 편향(Bias)

```python
def AND_Gate_Bias(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  result = np.sum(w*x) + b
  if result<= 0: 
    return 0
  elif result > 0:
    return 1

print(AND_Gate_Bias(0,0),AND_Gate_Bias(1,0),AND_Gate_Bias(0,1),AND_Gate_Bias(1,1), sep = ',')
```

AND 게이트 구현하기 : Boolean

```python
def AND(a, b): 
    if a == 1 and b == 1: 
        return True
    else: 
        return False
print(AND(0,0),AND(1,0),AND(0,1),AND(1,1), sep = ',')
```

NAND 게이트 구현하기

```python
def NAND_Gate(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5]) # AND GATE와 매개변수들의 부호만 반대! 
  b = 0.7
  result = np.sum(w*x) + b
  if result<= 0: 
    return 0
  elif result > 0:
    return 1

print(NAND_Gate(0,0),NAND_Gate(1,0),NAND_Gate(0,1),NAND_Gate(1,1), sep = ',')
```

NAND 게이트 구현하기 : Boolean

```python
def NAND(a, b): 
    if a == 1 and b == 1: 
        return False
    else: 
        return True

print(NAND(0,0),NAND(1,0),NAND(0,1),NAND(1,1), sep = ',')
```

OR 게이트 구현하기

```python
def OR_Gate(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5]) # AND GATE와 편향만 다르다!
  b = -0.2
  result = np.sum(w*x) + b
  if result<= 0: 
    return 0
  elif result > 0:
    return 1

print(OR_Gate(0,0),OR_Gate(1,0),OR_Gate(0,1),OR_Gate(1,1), sep = ',')
```

OR 게이트 구현하기 : Boolean

```python
def OR(a, b): 
    if a == 1: 
        return True
    elif b == 1: 
        return True
    else: 
        return False  

print(OR(0,0),OR(1,0),OR(0,1),OR(1,1), sep = ',')
```

XOR 게이트 구현하기

```python
def XOR_Gate(x1, x2):
  s1 = NAND_Gate(x1, x2)
  s2 = OR_Gate(x1, x2)
  y = AND_Gate_Bias(s1,s2)
  return y


print(XOR_Gate(0,0),XOR_Gate(1,0),XOR_Gate(0,1),XOR_Gate(1,1), sep = ',')
```

XOR 게이트 구현하기 : Boolean

```python
def XOR (a, b): 
    if a != b: 
        return True
    else: 
        return False
      
print(XOR(0,0),XOR(1,0),XOR(0,1),XOR(1,1), sep = ',')
```
