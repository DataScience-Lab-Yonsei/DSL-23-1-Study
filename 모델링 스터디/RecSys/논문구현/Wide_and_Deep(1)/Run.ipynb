```python
# 사용하는 데이터의 열이름을 담고 있다.

columns = ["age", "workclass", "fnlwgt", "education", "education_num",
                   "marital_status", "occupation", "relationship", "race", "gender",
                   "capital_gain", "capital_loss", "hours_per_week", "native_country",
                   "income_bracket"]
```

```python
# 사용할 데이터 불러오기

df_train = pd.read_csv(file_path + '/data/adult.data', sep=',', names = columns)
df_test = pd.read_csv(file_path + '/data/adult.test', sep=',', names = columns, skipinitialspace=True, skiprows=1)
df_train.head()
```

```python
# y값으로 사용할 'income_bracket(소득)'열 확인

df_train.income_bracket.value_counts()
```

```python
# y값을 label-encoding; 새로운 열 'income_label'
# 50k이하이면 -> 0
# 50k이상이면 -> 1

# y값이 (0,1)인 binary-classification 문제이다.

df_train['income_label'] = (df_train["income_bracket"].apply(lambda x: ">50K" in x)).astype(int)
df_test['income_label'] = (df_test["income_bracket"].apply(lambda x: ">50K" in x)).astype(int)
df_train.income_label.value_counts()
```

```python
# age열 확인 -> 그룹화를 하고자 한다.

df_train.age.hist()
```

```python
# age_group 추가

# 0~25(0), 25~65(1), 65~90(2)의 세 범주로 나눈다.

age_groups = [0, 25, 65, 90]
age_labels = range(len(age_groups) - 1)
df_train['age_group'] = pd.cut(df_train['age'], age_groups, labels=age_labels)
df_test['age_group'] = pd.cut(df_test['age'], age_groups, labels=age_labels)
df_train.age_group.value_counts()
```
